```python
from ultralytics import YOLO

# Loading pretrained YOLOv11  model
model = YOLO("yolo11n.pt")

# Training the model
model.train(
    data=data_yaml,
    epochs=150,
    batch=16,
    imgsz=900,
    name="pcb_detector",
    device=0
)
```

```python
from ultralytics import YOLO

# Loading trained model
model = YOLO("runs/detect/pcb_detector/weights/best.pt")

# Image paths
test_images = [
    "/content/drive/MyDrive/Project 3 Data/data/evaluation/ardmega.jpg",
    "/content/drive/MyDrive/Project 3 Data/data/evaluation/arduno.jpg",
    "/content/drive/MyDrive/Project 3 Data/data/evaluation/rasppi.jpg",
]

# Run predictions
for img in test_images:
    model.predict(img, save=True, conf=0.25)
    print("Finished:", img)
```
